```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Data Cleaning

```python
## missing values
# df.isnull().sum()
df[df.isnull().any(axis = 1)]
```

The dataset is converted into 2 sets based on Region from 122th index, we can make a new column based on the Region

1: "Bejaia Region Dataset"  
2: "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
df.loc[:122, "Region"] = 0
df.loc[122:, "Region"] = 1
```

```python
df.head()
```

```python
df.info()
```

```python
df[['Region']] = df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
##Removing the null values
df = df.dropna().reset_index(drop = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

#### Change the required columns as integer datatype


```python
df[['day','month','year','Temperature','RH','Ws']] = df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

#### Changing the other necessary columns to float data type

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
##Let's save the cleaned data set
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```
